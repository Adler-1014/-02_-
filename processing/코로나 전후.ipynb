```python
bus_station_lst = [
    station, n_entarance_1, elementary_1, h_apart_1, naksung_park_1,
    science_1, hoam_1, i_apart_1, hoam_2, family_r_1,
    s_back_door_1, undergraduate_residence_1, graduate_residence_1, dormitory_three_1, outdoor_hall,
    crf, heredity_lab_1, energy_lab, new_material_lab, power_plant,
    construction_lab, second_engineering, first_engineering, heredity_lab_2, professor_hall,
    gwanak_three, graduate_residence_2, undergraduate_residence_2, bk, family_r_2,
    science_2, naksung_park_2, h_apart_2, elementary_2, n_entarance_2
    ]

bus_station_name = [
    '낙성대역', '낙성대입구_1', '인헌초등학교_1', '낙성대현대아파트_1',
    '낙성대공원_1', '서울과학전시관_1', '호암교수회관_1', '인헌아파트_1',
    '호암교수회관_2', '가족생활동_1', '서울대후문_1', '학부생활관_1',
    '대학원생활관_1', '기숙사삼거리_1', '노천강당', '공동기기원',
    '유전공학연구_1', '에너지자원연구소', '신소재연구소', '제2파워플랜트',
    '건설환경종합연구소', '제2공학관', '제1공학관', '유전공학연구소_2',
    '교수회관입구', '관악사삼거리', '대학원생활관_2', '학부생활관_2',
    'BK국제관', '가족생활동_2', '서울과학전시관_2', '낙성대공원_2',
    '낙성대현대아파트_2', '인헌초등학교_2', '낙성대입구_2'
    ]
```

```python
year_lst = ['2018', '2019', '2020', '2021', '2022', '2023']
month_lst = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

# 승차
new_get_on_df = []

for bus_station in bus_station_lst:
  # 날짜 열을 문자열로 변환
  bus_station['date_str'] = bus_station['사용일자'].astype(str)
  bus_station_year = [[], [], [], [], [], []]

  for i in range(6):
    for j in range(12):
      year_month = year_lst[i] + month_lst[j]
      filtered_df = bus_station[bus_station['date_str'].str.startswith(year_month)]['승차총승객수'].sum()
      bus_station_year[i].append(int(filtered_df))

  new_get_on_df.append(bus_station_year)


# 하차
new_get_off_df = []

for bus_station in bus_station_lst:
  # 날짜 열을 문자열로 변환
  bus_station['date_str'] = bus_station['사용일자'].astype(str)
  bus_station_year = [[], [], [], [], [], []]

  for i in range(6):
    for j in range(12):
      year_month = year_lst[i] + month_lst[j]
      filtered_df = bus_station[bus_station['date_str'].str.startswith(year_month)]['하차총승객수'].sum()
      bus_station_year[i].append(int(filtered_df))

  new_get_off_df.append(bus_station_year)
```

```python
import plotly.express as px

# 모든 데이터 프레임을 합칠 빈 리스트 생성
all_get_on_data_long = []

for i in range(len(new_get_on_df)):
  data = {
  'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
  '2018': new_get_on_df[i][0],
  '2019': new_get_on_df[i][1],
  '2020': new_get_on_df[i][2],
  '2021': new_get_on_df[i][3],
  '2022': new_get_on_df[i][4],
  '2023': new_get_on_df[i][5],
  }
  # 데이터 프레임 생성
  get_on_data = pd.DataFrame(data)

  # 데이터 프레임을 길게 변환 (Wide-to-Long)
  get_on_data_long = get_on_data.melt(id_vars='Month', var_name='Year', value_name='Value')

  station_name = bus_station_name[i]

  # 역명 추가
  get_on_data_long['Station'] = bus_station_name[i]

  # 생성된 데이터 프레임을 빈 리스트에 추가
  all_get_on_data_long.append(get_on_data_long)

  # 라인 플롯 그리기
  fig = px.line(
      get_on_data_long,
      x = 'Month',  # x축
      y = 'Value',  # y축
      color = 'Year',  # 연도별로 색상 구분
      title = f'{station_name}',  # f-스트링 수정
      labels = {'Month': '월', 'Value': '값', 'Year': '연도'}
  )

  fig.show()
```

```python
# 하차
import plotly.express as px

# 모든 데이터 프레임을 합칠 빈 리스트 생성
all_get_off_data_long = []

for i in range(len(new_get_off_df)):
  data = {
  'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
  '2018': new_get_off_df[i][0],
  '2019': new_get_off_df[i][1],
  '2020': new_get_off_df[i][2],
  '2021': new_get_off_df[i][3],
  '2022': new_get_off_df[i][4],
  '2023': new_get_off_df[i][5],
  }
  # 데이터 프레임 생성
  get_off_data = pd.DataFrame(data)

  # 데이터 프레임을 길게 변환 (Wide-to-Long)
  get_off_data_long = get_off_data.melt(id_vars='Month', var_name='Year', value_name='Value')

  station_name = bus_station_name[i]

  # 역명 추가
  get_off_data_long['Station'] = bus_station_name[i]

  # 생성된 데이터 프레임을 빈 리스트에 추가
  all_get_off_data_long.append(get_off_data_long)

  # 라인 플롯 그리기
  fig = px.line(
      get_off_data_long,
      x = 'Month',  # x축
      y = 'Value',  # y축
      color = 'Year',  # 연도별로 색상 구분
      title = f'{station_name}',  # f-스트링 수정
      labels = {'Month': '월', 'Value': '값', 'Year': '연도'}
  )

  fig.show()
```
