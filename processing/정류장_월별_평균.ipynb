```python
# 전체(각 정류장에 따른 시간별 전체 이용자 수, 2018년 ~ 2023년 데이터)
# 승차
passenger_get_on_sum_lst = []
for i in range(35):
  passenger_get_on_sum = [bus_station_lst[i][get_on_time_lst[j]].sum() for j in range(20)]
  passenger_get_on_sum_lst.append(passenger_get_on_sum)

# 하차
passenger_get_off_sum_lst = []
for i in range(35):
  passenger_get_off_sum = [bus_station_lst[i][get_off_time_lst[j]].sum() for j in range(20)]
  passenger_get_off_sum_lst.append(passenger_get_off_sum)

print(passenger_get_off_sum_lst)

len(passenger_get_off_sum_lst)
```

```python
# 전체(각 정류장 평균 이용자 수, 2018년 ~ 2023년 데이터)
bus_station_get_on_total_sum = []
for i in range(len(bus_station_lst)):
  total = 0
  for j in range(20):
    total += passenger_get_on_sum_lst[i][j]
  bus_station_get_on_total_sum.append(total // (365 * 6))


bus_station_get_off_total_sum = []
for i in range(len(bus_station_lst)):
  total = 0
  for j in range(20):
    total += passenger_get_off_sum_lst[i][j]
  bus_station_get_off_total_sum.append(total // (365 * 6))

bus_station_get_off_total_sum
```
